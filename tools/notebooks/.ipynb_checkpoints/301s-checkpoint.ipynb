```python
import csv
import requests
import re
from lxml import etree as ET
```

```python
tree = ET.parse('../data/kind-catalog.xml')
root = tree.getroot()

CATALOG = {
    "categories": [],
    "products": []
}

VARIANT_MASTER_MAP = {}

for child in root:
    if child.tag == "product":
        master_id = child.attrib["product-id"]
        CATALOG["products"].append(master_id)
        
        for children in child:
            if children.tag == 'variations':
                for variants in children[1]:
                    variant_id = variants.attrib['product-id']
                    VARIANT_MASTER_MAP[variant_id] = master_id
                    
storefront_tree = ET.parse('../data/kind-storefront.xml')
storefront_root = tree.getroot()

for child in root:
    if child.tag == "category":
        CATALOG["categories"].append(child.attrib["category-id"])
    
reader = csv.DictReader(open('../data/redirects.csv', 'rt', encoding="utf-8-sig"))
dict_urls = []
for line in reader:
    dict_urls.append(line)
    
reader = csv.DictReader(open('../data/product-redirects.csv', 'rt', encoding="utf-8-sig"))
dict_product = []
for line in reader:
    dict_product.append(line)
```

```python
REMAINING_URLS = []
STATIC_REDIRECTS = []
count = 0
PRODUCT_IDS = []
FOUND_PRODUCTS = []
PRODUCT_ID = ''
STATIC_REDIRECTS = []
FOUND_URLS = []
NOT_FOUND_URLS = []

for product in dict_product:
    FOUND_PRODUCT = False
    PRODUCT_ID = product['new_product_id']

    for url in dict_urls:
        PRODUCT_URL = url['urls'].rsplit('/', 1)[-1]
        if PRODUCT_URL == product['magento_url']:
            count += 1
            FOUND_PRODUCTS.append(PRODUCT_ID)
            FOUND_URLS.append(url['urls'])
            FOUND_PRODUCT = True
            static_url = url['urls'].replace('www.kindsnacks.com/', '')
            STATIC_REDIRECTS.append('**' + static_url + '** ' + 'p,,,Product-Show,,pid,' + PRODUCT_ID)
            
    if not FOUND_PRODUCT and PRODUCT_ID not in FOUND_PRODUCTS:
        STATIC_REDIRECTS.append('**' + product['magento_url'] + '** ' + 'p,,,Product-Show,,pid,' + PRODUCT_ID)
        
for url in dict_urls:
    FOUND_PRODUCT = False
    PRODUCT_URL = url['urls'].rsplit('/', 1)[-1]
    
    for product in dict_product:
        if PRODUCT_URL == product['magento_url']:
            FOUND_PRODUCT = True
            break
            
    if not FOUND_PRODUCT:
        NOT_FOUND_URLS.append({
            "source": url["source"],
            "urls": url["urls"],
            "redirect_to": url["redirect_to"]
        })
```

```python
with open('../data/product_static_redirects.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for url in STATIC_REDIRECTS:
        writer.writerow([url])
```

```python
REMAINING_URLS = []
ACTUAL_URLS = []
STATIC_REDIRECTS = []
NO_CONNECT_URLS = []
TOTAL = 0
FOUND = 0
PAGE_NOT_FOUND = []
PRODUCT_IDS = []
PRODUCT_IDS_URLS = []

for url in NOT_FOUND_URLS:
    types = r"blog|press-releases|media-center|news|contact-us|foundation|ambassador"
    match_string = re.search(types, url["urls"])
    if match_string == None:
        try:
            regex = r"data-id=\"(.*?)\""
            catalog_r = r"catalog-category-view"
            prouduct_r = r"catalog-product-view"
            home_r = r"cms-home"

            request_url = 'https://' + url["urls"]
            response = requests.get(request_url)
            HTML = response.text

            match_catalog = re.search(catalog_r, HTML)
            match_product = re.search(prouduct_r, HTML)
            match_home = re.search(home_r, HTML)

            foundMatch = False
            foundProduct = False
            foundCatalog = False
            foundHome = False
            foundUnknownProduct = False

            PRODUCT_ID = ''
            CATEGORY_ID = ''

            if match_catalog:
                CATEGORY_ID = request_url.rsplit('/', 1)[-1]
                if CATEGORY_ID in CATALOG["categories"]:
                    foundMatch = True
                    foundCatalog = True

            elif match_product:
                match_id = re.search(regex, HTML)
                if match_id:
                    PRODUCT_ID = match_id.group(1)
                    if PRODUCT_ID in VARIANT_MASTER_MAP:
                        PRODUCT_ID = VARIANT_MASTER_MAP[PRODUCT_ID]
                        foundMatch = True
                        foundProduct = True
                    else:
                        PRODUCT_IDS.append(match_id)
                        foundMatch = True
                        foundUnknownProduct = True
                        

            elif match_home:
                foundMatch = True
                foundHome = True

            if foundMatch:
                static_url = request_url.replace('https://www.kindsnacks.com/', '')
                static_mapping = ''
                FOUND += 1

                if foundUnknownProduct:
                    PRODUCT_IDS_URLS.append({
                        "source": url["source"],
                        "urls": url["urls"],
                        "redirect_to": url["redirect_to"]
                    })
                if foundProduct:
                    STATIC_REDIRECTS.append(static_url + ' ' + 'p,,,Product-Show,,pid,' + PRODUCT_ID)
                if foundCatalog:
                    STATIC_REDIRECTS.append(static_url + ' ' + 'p,,,Search-Show,,cgid,' + CATEGORY_ID)
                if foundHome:
                    STATIC_REDIRECTS.append(static_url + ' ' + 'p,,,Home-Show,')

            elif response.status_code == 404:
                PAGE_NOT_FOUND.append({
                    "source": url["source"],
                    "urls": url["urls"],
                    "redirect_to": url["redirect_to"]
                })
                FOUND += 1

            else:
                REMAINING_URLS.append({
                    "source": url["source"],
                    "urls": url["urls"],
                    "redirect_to": url["redirect_to"]
                })
                
            if response.status_code != 404:
                ACTUAL_URLS.append({
                    "source": url["source"],
                    "urls": url["urls"],
                    "redirect_to": url["redirect_to"]
                })
                
        except:
            NO_CONNECT_URLS.append({
                "source": url["source"],
                "urls": url["urls"],
                "redirect_to": url["redirect_to"]
            })
```

```python
REMAINING_URLS = []
STATIC_REDIRECTS = []
NO_CONNECT_URLS = []
TOTAL = 0
FOUND = 0
PAGE_NOT_FOUND = []

for url in dict_urls:
    try:
        regex = r"data-id=\"(.*?)\""
        catalog_r = r"catalog-category-view"
        prouduct_r = r"catalog-product-view"
        home_r = r"cms-home"

        request_url = 'https://' + url["urls"]
        response = requests.get(request_url)
        HTML = response.text

        match_catalog = re.search(catalog_r, HTML)
        match_product = re.search(prouduct_r, HTML)
        match_home = re.search(home_r, HTML)

        foundMatch = False
        foundProduct = False
        foundCatalog = False
        foundHome = False

        PRODUCT_ID = ''
        CATEGORY_ID = ''

        if match_catalog:
            CATEGORY_ID = request_url.rsplit('/', 1)[-1]
            if CATEGORY_ID in CATALOG["categories"]:
                foundMatch = True
                foundCatalog = True

        elif match_product:
            match_id = re.search(regex, HTML)
            if match_id:
                PRODUCT_ID = match_id.group(1)
                if PRODUCT_ID in CATALOG["products"]:
                    foundMatch = True
                    foundProduct = True

        elif match_home:
            foundMatch = True
            foundHome = True

        if foundMatch:
            static_url = request_url.replace('https://www.kindsnacks.com/', '')
            static_mapping = ''
            FOUND += 1

            if foundProduct:
                STATIC_REDIRECTS.append(static_url + ' ' + 'p,,,Product-Show,,pid,' + PRODUCT_ID)
            if foundCatalog:
                STATIC_REDIRECTS.append(static_url + ' ' + 'p,,,Search-Show,,cgid,' + CATEGORY_ID)
            if foundHome:
                STATIC_REDIRECTS.append(static_url + ' ' + 'p,,,Home-Show,')


        elif response.status_code == 404:
            PAGE_NOT_FOUND.append({
                "source": url["source"],
                "urls": url["urls"],
                "redirect_to": url["redirect_to"]
            })
            FOUND += 1

        else:
            REMAINING_URLS.append({
                "source": url["source"],
                "urls": url["urls"],
                "redirect_to": url["redirect_to"]
            })
    except:
        NO_CONNECT_URLS.append({
            "source": url["source"],
            "urls": url["urls"],
            "redirect_to": url["redirect_to"]
        })

    TOTAL += 1
    print('FOUND: ' + str(FOUND))
    print('TOTAL: ' + str(TOTAL))
    print('\n\n')
```

```python
print(len(STATIC_REDIRECTS))
print(len(REMAINING_URLS))
print(len(PAGE_NOT_FOUND))

with open('../data/remaining_urls.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for url in REMAINING_URLS:
        writer.writerow([url["source"], url["urls"], url["redirect_to"]])
        
with open('../data/static_redirects.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for url in STATIC_REDIRECTS:
        writer.writerow([url])
```
